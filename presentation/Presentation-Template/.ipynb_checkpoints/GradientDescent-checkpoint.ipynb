```python
import os
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
def doGradientDescent(case):
    
    if case == 'normal':
        rate = 0.25
        first_x = 8
    elif case =='slow':
        rate = 0.06
        first_x = 8
    elif case =='fast':
        rate=1.1
        first_x = 8
    elif case == 'oscillations':
        rate = 1
        first_x = 6
    
    
    next_x = 6  # We start the search at x=6
    gamma = 0.01  # Step size multiplier
    precision = 0.001
    max_iterations = 100


    step_size = 1
    counter = 0

    # Derivative function
    x = np.linspace(0, 10, 1000)
    y = (x - 5)**2

    f = lambda x: (x-5)**2
    df = lambda x: 2*(x - 5)

    z = np.linspace(6, 10, 100)
    derivative = 6
    w = 6*z - 39

    def gradient(x):
        return 2.0*(x-6)


    fig = plt.figure(figsize=(8, 8))
    ax = plt.gca()

    plt.plot(x,y, color = 'blue', lw=3, label=r'$y = (x - 5)^{2}$')

    ###########################################################################    
    cur_x = 8;
    cur_y = f(cur_x)
    next_x = 8

    first_y = f(first_x)
    first_gradient = df(first_x)
    second_x = first_x - rate * first_gradient
    second_y = f(second_x)

    if case != 'oscillations':
        plt.plot(z, w, color = 'magenta', lw=1, label='Gradient')
        plt.text(9, 2.8,'Gradient',fontsize=14)
        ax.arrow(8.8, 3, -1.3, 0, head_width=0.1, head_length=0.3, fc='blue', ec='red')
    else:
        ax.arrow(second_x+0.25, second_y, first_x -second_x -0.75 , 0,
                 head_width=0.1,head_length=0.3, fc='blue', ec='red')
        ax.arrow(first_x -0.25, second_y+0.25, second_x - first_x + 0.75, 0,
                 head_width=0.1,head_length=0.3, fc='blue', ec='green')
        plt.text(4.25, 1.5, 'Oscillations', fontsize=20)

    plt.scatter(first_x, first_y,s=100,color='red',label='current y')
    plt.plot([first_x, first_x],[0,first_y], color='black', lw = 2)

    plt.scatter(second_x, second_y, s=100,color='green',label='next_y')
    plt.plot([second_x, second_x],[0,second_y], color='black', lw = 2)

    plt.scatter(first_x, 0, s=50, color='red')
    plt.text(first_x + 0.25, 0.3,r'$x_{prev}$', color='red', fontsize=14)

    plt.scatter(second_x, 0, color='green', s=50)
    plt.text(second_x + 0.2, 0.3,r'$x_{new}$', color='green', fontsize=14)


    # for oscillations take rate = 1 and next_x = 6
    ##########
    print("Learning Rate is:" , rate)
    next_x = first_x
    for i in range(max_iterations):
        cur_x = next_x
        next_x = cur_x - rate * df(cur_x)
        step = next_x - cur_x
        print("Iteration: ", i+1, " x=", round(next_x, 4)) 
        if abs(step) <= precision:
            break

    print("Minimum at", round(next_x, 4), ' after ', i+1, ' iterations')


    plt.legend(loc='upper center')
    if case == 'oscillations':
        plt.xlim(3, 7)
        plt.ylim(0,4) 
        plt.xticks([3,4,5,6,7])
        plt.yticks([0,1,2,3,4])

    else:
        plt.xlim(0,12)
        plt.ylim(0,13)
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')

    s_title = 'start x=' + str(first_x)  + ', learning rate: ' + str(rate)
    plt.title(s_title)

    plt.savefig(case + ".pdf", bbox_inches='tight')
    plt.show()
```

```python
doGradientDescent(case='normal' )
```

```python
doGradientDescent(case = 'slow')
```

```python
doGradientDescent(case = 'fast')
```

```python
doGradientDescent(case = 'oscillations')
```
