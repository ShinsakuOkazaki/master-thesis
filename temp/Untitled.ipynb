```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
initials = [10, 100, 1000, 10000]
finals = [ 99, 999, 9999]

json_files = {}
for i in initials:
    for j in finals:
        test = "intial_{}_fanal_{}".format(i, j)
        path = test + "/base/estimates.json"
        with open("criterion/" + path, "r") as f:
            json_files[(i, j)] = json.load(f)
```

```python
estimated999 = {}
for i in initials:
    estimated999[(i, 999)] = json_files[(i, 999)]["Slope"]['point_estimate']
```

```python
estimated999
```

```python
time = []
for i in initials:
    time.append(estimated999[(i, 999)])
```

```python
time
```

```python
plt.figure(figsize = (10, 8))
plt.plot(initials, time)
plt.show()
```
