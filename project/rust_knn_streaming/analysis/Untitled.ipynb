```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfs = {}
for m in ['deep', 'arc']:
    for s in ['s1', 's2']:       
        with open("ganglia-b3-" + s + "-" + m + ".csv", "r") as f:
            file = [s.strip().split(",") for s in f.readlines()]
            df = pd.DataFrame(file)
            header = df.iloc[0]

            df = df[1:]
            df.columns = header
            type_map = {"Timestamp": "datetime64[ns]"}
            for h in header[1:]:
                type_map[h] = "float"
            df = df.astype(type_map)
            dfs[m + "-" + s] = df
```
