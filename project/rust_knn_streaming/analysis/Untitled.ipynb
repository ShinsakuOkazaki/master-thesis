```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
files = []
for r, d, f in os.walk("../profile/"):
    for file in f:
        files.append(os.path.join("../profile/", file))
```

```python
methods = []
n_batchs = []
ks = []
n_neighborses = []
object_list = []
for file in files:
    with open(file, 'r') as f:
        content = f.readlines()
        if (len(content) == 0):
            continue
        items = file.split('/')[2].split("_")
        methods.append(int(items[0][6]))
        n_batchs.append(int(items[1][6]))
        ks.append(int(items[2][1:]))
        n_neighborses.append(int(items[3][10:12]))
        dic = {}
        for line in content:
            size, ob = line.split("#")
            ob = str(ob[:-1])
            print(ob)
            if (ob in dic.keys()):
                dic[ob].append(int(size))
            else:
                dic[ob] = [int(size)]
        objects = {}
        for ob, sizes in dic.items():
            objects[ob] = np.mean(sizes)
        object_list.append(objects)
```

```python
obj_batch = {}
for k in np.unique(n_batchs):
    obj_batch[k] = []
for i in range(len(n_batchs)):
    obj_batch[n_batchs[i]].append(object_list[i])
```

```python
def map_obj_list(l):
    obj_dict = {}
    for ob in l:
        for k, v in ob.items():
            if(k in obj_dict):
                obj_dict[k].append(v)
            else:
                obj_dict[k] = [v]
    return obj_dict
```

```python
for k in obj_batch.keys():
    obj_batch[k] = list(map(lambda x: map_obj_list(x), obj_batch[k]))
```

```python
conf_dict = {"method": methods, "n_batch": n_batchs, "k": ks, "n_neighbors": n_neighborses}
df = pd.DataFrame({**conf_dict, **obj_dict})
```
